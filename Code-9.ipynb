# SOBEL

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('waifu.jpg')
```

```python
sobelx = cv2.Sobel(img, cv2.CV_8U, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_8U, 0, 1, ksize=5)
sobel = sobelx + sobely
```

```python
fig, axes = plt.subplots(4, 2, figsize=(20, 20))
ax = axes.ravel()

ax[0].imshow(img, cmap='gray')
ax[0].set_title("input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title("histogram")

ax[2].imshow(sobelx, cmap='gray')
ax[2].set_title("output")
ax[3].hist(img.ravel(), bins=256)
ax[3].set_title("histogram")

ax[4].imshow(sobely, cmap='gray')
ax[4].set_title("output")
ax[5].hist(img.ravel(), bins=256)
ax[5].set_title("histogram")

ax[6].imshow(sobel, cmap='gray')
ax[6].set_title("output")
ax[7].hist(img.ravel(), bins=256)
ax[7].set_title("histogram")

fig.tight_layout()
```

# CANNY

```python
canny = cv2.Canny(img,100,200)
```
